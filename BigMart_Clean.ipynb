```python
#Replacing the nan values with the mean weight
BigMart_TrainData['Item_Weight']= BigMart_TrainData['Item_Weight'].fillna(BigMart_TrainData['Item_Weight'].mean())
BigMart_TestData['Item_Weight']= BigMart_TestData['Item_Weight'].fillna(BigMart_TestData['Item_Weight'].mean())
```

```python
#Copying the datasets into new data frames for imputing
BigMart_TrainData_impute = BigMart_TrainData.copy(deep=True)
BigMart_TestData_impute = BigMart_TestData.copy(deep=True)
```

```python
#Imputing the test dataset with the most frequent Outlet size
impute_size = SimpleImputer(missing_values = np.nan, strategy="most_frequent") 
BigMart_TrainData_impute['Outlet_Size']=impute_size.fit_transform(BigMart_TrainData_impute[['Outlet_Size']])
BigMart_TrainData_impute
```

```python
#Imputing the test dataset with the most frequent Outlet size
impute_size = SimpleImputer(missing_values = np.nan, strategy="most_frequent") 
BigMart_TestData_impute['Outlet_Size']=impute_size.fit_transform(BigMart_TestData_impute[['Outlet_Size']])
BigMart_TestData_impute
```

```python
#Saving the datasets into new csv files 
BigMart_TrainData_impute.to_csv('Train_Cleaned.csv')
BigMart_TestData_impute.to_csv('Test_Cleaned.csv')
```
